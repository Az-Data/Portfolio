# My Life as Data

I wanted to create a few simple visualization to flex my python muscles. I decided to quantify and plot various aspects of my life. Much of the data is vague estimation and some of the data is completely fabricated for entertainment value. Please enjoy My life as Data. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
import numpy as np
sns.set(style="white")
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

My university years were a good bunch of years, though if I were to simplify it, it could measure my sobriety in a single stacked bar chart.

```python
time = {'Sober': 80, 'Not Sober': 20}

df = pd.DataFrame(list(time.items()), columns=['Sobriety', '% of time'])

df.set_index('Sobriety').T.plot(kind='bar', stacked=True)
sns.despine(top=True, right=True, left=True, bottom=True)

plt.show()
```

```python
df2 = pd.DataFrame({'Sleep': 8, 
                    'Workout': 0.75, 
                    'Eating/Cooking': 2, 
                    'Reading': 1.5, 
                    'Data Science': 4, 
                    'Work': 2, 
                    'Public Speaking': 0.25, 
                    'Shopping':0.25, 
                    'Friends':1, 
                    'Media': 1, 
                    'Procrastinating':3.25}, index=[''])
df2 = df2[['Sleep', 'Workout', 'Eating/Cooking', 'Reading', 'Data Science', 'Work', 'Public Speaking', 'Shopping', 'Friends', 'Media', 'Procrastinating']]
#display(df2)
```

```python
ax = df2.plot.barh(stacked=True, colormap=plt.cm.RdYlGn)
ax.legend(loc='upper center', ncol=3, fontsize='x-large')
plt.xlabel('Hours', fontsize=14)
sns.despine(top=True, right=True, left=True, bottom=True)

#plt.savefig('foo.png')
plt.show()
```

```python
dict1 = {'Rom Com': 10, 'Comedy':25, 'Action': 40, 'Garbage': 20, 'Drama': 5}

watch = pd.DataFrame(dict1, index=[''])

watch
```

```python
watch.plot.bar(stacked=True, colormap=plt.cm.RdBu)
sns.despine(top=True, right=True, left=True, bottom=True)
plt.legend(fontsize='x-large')
plt.ylabel('% of the Time', fontsize=14)
plt.xlabel('Staying in', fontsize=14)
plt.show()
```

```python
watch2 = pd.DataFrame({'Anything with lots of CGI':100}, index=[''])

watch2
```

```python
watch2.plot.bar()
plt.ylabel('% of the time', fontsize=14)
plt.xlabel('At the Cinema', fontsize=14)
plt.legend(fontsize='x-large')
sns.despine(top=True, right=True, left=True, bottom=True)

plt.show()
```

```python
# People I see on a weekly basis

people = pd.DataFrame({'Family': 40, 'Co-workers': 30, 'Friends': 25, 'Aquaintences': 50, 'Toastmasters':30}, index=[''])

people
```

```python
plt.figure()

groups = ['Family', 'Co-workers', 'Friends', 'Aquaintences', 'Toastmasters']
pos = np.arange(len(groups))
people = [30, 30, 25, 50, 30]

bars = plt.bar(pos, people, align='center', linewidth=0, color='lightslategrey', yerr=[2,5,5,10,4], capsize=5)
bars[0].set_color('#1F77B4')

plt.xticks(pos, groups, fontsize=14)
plt.ylabel('No. of people', fontsize=15)

sns.despine(top=True, right=True, left=True, bottom=True)
plt.show()
```
